## Bar charts

```python
google = data.DataReader(name='GOOG', data_source='google', start='2004-01-01',end='2016-12-12')
google
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use('ggplot')
google["Close"].apply(rank_performance).value_counts().plot(kind='barh')
```

## pie chart

usuful for showing how seperate data makes a whole

```python
apple = data.DataReader(name='AAPL', data_source='google', start='2012-01-01', end='2016-12-12')
apple.head(3)
```

```python
apple['Close'].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 93.12191967871486:
        return "Above average"
    else:
        return "Below average"
```

```python
apple['Close'].apply(rank_performance).value_counts().plot(kind='pie', legend=True)
```

## Histogram

good at showing distributions of data

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
custom_round(535.6)
```

```python
google['High'].apply(custom_round).value_counts().sort_index()
```

```python
google['High'].apply(custom_round).nunique()
```

```python
google['High'].apply(custom_round).plot(kind='hist', bins=9)
```
